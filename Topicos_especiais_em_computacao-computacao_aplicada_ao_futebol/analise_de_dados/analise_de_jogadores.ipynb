```python
import pandas as pd
import numpy as np

# Ler o dataset base
df_base = pd.read_csv("dataset_base.csv")

# Calcular diferenças de posição
dx_A = df_base["x_A"].diff()
dy_A = df_base["y_A"].diff()

dx_B = df_base["x_B"].diff()
dy_B = df_base["y_B"].diff()

# Intervalo de tempo
dt = df_base["tempo"].diff()

# Calcular velocidade
velocidade_A = np.sqrt(dx_A**2 + dy_A**2) / dt
velocidade_B = np.sqrt(dx_B**2 + dy_B**2) / dt

df_base["velocidade_A"] = velocidade_A
df_base["velocidade_B"] = velocidade_B

# Calcular aceleração
aceleracao_A = df_base["velocidade_A"].diff() / dt
aceleracao_B = df_base["velocidade_B"].diff() / dt

df_base["aceleracao_A"] = aceleracao_A
df_base["aceleracao_B"] = aceleracao_B

# 1. Velocidade máxima
vel_max_A = df_base["velocidade_A"].max()
vel_max_B = df_base["velocidade_B"].max()

# 2. Distância total percorrida
dist_total_A = np.sqrt(dx_A**2 + dy_A**2).sum()
dist_total_B = np.sqrt(dx_B**2 + dy_B**2).sum()

# 3. Tempo em alta intensidade
# Critério: velocidade > 6 m/s
tempo_alta_A = (df_base["velocidade_A"] > 6).sum() * dt.mean()
tempo_alta_B = (df_base["velocidade_B"] > 6).sum() * dt.mean()

# 4. Número de acelerações relevantes
# Critério: aceleração > 0.5 m/s²
acel_relev_A = (df_base["aceleracao_A"] > 0.5).sum()
acel_relev_B = (df_base["aceleracao_B"] > 0.5).sum()

# Mostrar resultados
print("=== ITEM 4 - COMPARAÇÃO DE MÉTRICAS ===")
print(f"Velocidade máxima do jogador A: {vel_max_A:.2f} m/s")
print(f"Velocidade máxima do jogador B: {vel_max_B:.2f} m/s")

print(f"Distância total percorrida pelo jogador A: {dist_total_A:.2f} m")
print(f"Distância total percorrida pelo jogador B: {dist_total_B:.2f} m")

print(f"Tempo em alta intensidade do jogador A (v > 6 m/s): {tempo_alta_A:.2f} s")
print(f"Tempo em alta intensidade do jogador B (v > 6 m/s): {tempo_alta_B:.2f} s")

print(f"Número de acelerações relevantes do jogador A (a > 0.5 m/s²): {acel_relev_A}")
print(f"Número de acelerações relevantes do jogador B (a > 0.5 m/s²): {acel_relev_B}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df_alta = pd.read_csv("dataset_alta_freq.csv")
df_base = pd.read_csv("dataset_base.csv")
df_ruido = pd.read_csv("dataset_ruido.csv")


dx_A = df_base["x_A"].diff()
dy_A = df_base["y_A"].diff()


dx_B = df_base["x_B"].diff()
dy_B = df_base["y_B"].diff()


dt = df_base["tempo"].diff()


velocidade_A = np.sqrt(dx_A**2 + dy_A**2) / dt
velocidade_B = np.sqrt(dx_B**2 + dy_B**2) / dt


df_base["velocidade_A"] = velocidade_A
df_base["velocidade_B"] = velocidade_B

velocidade_A_suavizada = (velocidade_A.shift(-1) + velocidade_A + velocidade_A.shift(1)) / 3
velocidade_B_suavizada = (velocidade_B.shift(-1) + velocidade_B + velocidade_B.shift(1)) / 3

df_base["velocidade_A_suavizada"] = velocidade_A_suavizada
df_base["velocidade_B_suavizada"] = velocidade_B_suavizada


print("Velocidades normais: ")
print(velocidade_A)
print(velocidade_B)
print("Velocidades suavizadas: ")
print(velocidade_A_suavizada)
print(velocidade_B_suavizada)

plt.plot(df_base["tempo"], df_base["velocidade_A"], label="Jogador A")
plt.plot(df_base["tempo"], df_base["velocidade_B"], label="Jogador B")


plt.xlabel("Tempo")
plt.ylabel("Velocidade")
plt.title("Velocidade x Tempo")
plt.legend()


plt.show()

plt.plot(df_base["tempo"], df_base["velocidade_A_suavizada"], label="Jogador A")
plt.plot(df_base["tempo"], df_base["velocidade_B_suavizada"], label="Jogador B")


plt.xlabel("Tempo")
plt.ylabel("Velocidade suavizada")
plt.title("Velocidade suavizada x Tempo")
plt.legend()


plt.show()
```
